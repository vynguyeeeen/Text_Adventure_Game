```python
"""
Completed Sun Oct 18 2020 

@author: phuong-vy.nguyen 

"""

# Docstring - provide overview of the game and identify known errors 
"""
DocString:
    
A)  INTRODUCTION:

    The following game, titled ‘Dare to Dream’ (D2D) turns Walt Disney's 
    infamous quote into existence: “if you can dream it, you can do it”. 

    The game follows a brand new cryosleep program that will help make 
    anyone’s wildest, supposedly unreachable, dreams come true. 

    Let’s say yours was to travel to the legendary lost empire of Atlantis. 
    D2D will get you there. All it takes is your imagination, occasional 
    logic and the ability to successfully complete 4 rounds: 

        ROUND 1: Confirm interests in participating
        ROUND 2: Recruit a teammate
        ROUND 3: Pick a mode of transportation 
        ROUND 4: Pass security checkpoints 

    Good luck. Good night. And don’t let the bed bug bite. 

B)  KNOWN BUGS AND/OR ERRORS: 
    None.

"""

# Define time function to add delay execution in the game 
import time

# Define exit to quit the game 
from sys import exit


# Define the function for of the beginning of the game 
def start():
    
    print("Guess what?\n")
    
    # importing time to use time.sleep() for better user experience 
    time.sleep(2) 
    
    print("Today's your lucky day!\n") 
    time.sleep(3) 
    
    print(f""" 
Word on the streets is you've been wanting to visit the underwater... 
legendary... empire of Atlantis!!\n""") 
    time.sleep(7)
    
    print("We'll help you get there... How? SLEEP.\n")
    time.sleep(5)
    
    print(f"""
Our brand new technology - the cryosleep chamber - will help link your 
consciousness to an avatar made identical to yourself. Together, you and 
your avatar, will go on a mission to make the impossible, possible!
""")
    time.sleep(13)
    
    # call room 1 (name Intro) to start the 1st stage of the game  
    intro()


# Define the function of room 1  
def intro():
    
    # prompt user to input their interest: answer 'yes' or 'no'
    print(f"""\nAre you ready?! 
    1. Yes
    2. No
""")
    
    # propmt user to enter an answer for question: 'are you ready?'
    choice = (input(prompt='>> '))
    
    # control user input by returning a lowercased or original string
    choice = choice.lower()
    
    # start conditional statement for question: 'are you ready?'
    if "1" in choice or "yes" in choice:
        print(f"""
That's the spirit! Lay down. Sink into those pillows...
""")
        # time delay for suspense 
        time.sleep(3)
        
        input(prompt='When you\'re ready... <press Enter to wake up>\n')
        print('ZZzz...' * 8, end = '  ᕕ( ᐛ )ᕗ') 
        time.sleep(2)
        
        # provide user clear instructions for the entire game 
        print(f"""\n
Your mission, should you choose to accept it, involves being a CIA 
cryptologist. You recently discovered the whereabouts of Atlantis, 
but you must see it to believe it.

To see it - in chronological order, you must complete the following:
    1. Recruit a teammate to accompany you on this trip 
    2. Pick a fitting mode of transportation
    3. Provide the correct answers to pass security checkpoints 
""")  
        
        # prompt users to continue playing 
        input(prompt='<Press Enter to continue>\n')
        
        print("...This message will self-destruct in five seconds...\n")
        
        # time delay for suspense 
        time.sleep(4)
        print('5')
        time.sleep(1)
        print('4')
        time.sleep(1)
        print('3')
        time.sleep(1)
        print('2')
        time.sleep(1)
        print('1')
        time.sleep(1)
        print(f"""   
  （ • •） 
╔uu═══════════════════╗
 < MESSAGE DESTROYED > 
╚═════════════════════╝
""")
        
        # prompt users to continue playing 
        input(prompt='<Press Enter to continue>\n')
        
        # introduce the next stage if user answered 'yes'
        print(f"""

┌──────────── ⋆⋅☆⋅⋆ ───────────┐
  TO BEGIN: RECRUIT A TEAMMATE
└──────────── ⋆⋅☆⋅⋆ ───────────┘
""")
        
        # call for room 2 (name Recruit) if user answered 'yes'
        recruit()
    
    elif "2" in choice or "no" in choice:
        
        # prompt user to reenter an input 
        print("\nC'mon, give it another try!")
        
        # call for room 1 - endless tries until user answer 'yes'
        intro()
        
    else:
        
        # prompt user to reenter a valid input - provid clearer instructions
        print("\nInvalid. Try 'yes' or 'no'.")
        
        # call for room 1 - endless tries until user answer 'yes'
        intro()
    
    # End conditional statement for question: 'are you ready?'

    
# Define the function of room 2         
def recruit():
    
    # print question for room 2 
    print(f"""From the following options: who would you like to go with?
    1. Steal, a former cop in charge of theft 
    2. Aqua, a marine biologist 
    3. Dough, an entrepreneur with cash 
""")

    # propmt user to enter the answer for teammate question 
    # introduce a variable 
    teammate = (input(prompt='>> '))
    
    # control user input by returning a lowercased or original string
    teammate = teammate.lower()
    
    # start conditional statement for teammate question 
    if "1" in teammate or "steal" in teammate or \
    "cop" in teammate or "theft" in teammate: 
        print(f"""\nSorry, Steal has opted for a retirement trip to the
Bahamas.\n""")
        
        # call for room 2 - endless tries until correct answer is inputted
        recruit()
    
    elif "2" in teammate or "aqua" in teammate or \
    "biologist" in teammate:
        print("\nWise choice!\n")
        
        # prompt user to continue playing when answer is correct 
        input(prompt='\n<Press Enter to continue>\n') 
        
        # print instructions for the next stage when answer is correct 
        print(f"""

┌─────────────── ⋆⋅☆⋅⋆ ──────────────┐
  TO CONTINUE: PICK A TRANSPORTATION
└─────────────── ⋆⋅☆⋅⋆ ──────────────┘

Uh oh, the gate to Atlantis is closing soon. There's only ONE way there. 
You have TWO CHANCES to decide. Otherwise, your trip ends here...
""")
        
        # call for room 3 (name Travel) when answer is correct 
        travel()
    
    elif "3" in teammate or "dough" in teammate or \
    "entrepreneur" in teammate or "cash" in teammate: 
        print('\nHe\'s kind of a weirdough. You don\'t want him.\n')
        
        # call for room 2 - endless tries until correct answer is inputted
        recruit()
    
    else:
        # print instructions to control user input 
        print('\nInvalid entry. Try picking someone from the list.')
        
        # call for room 2 - endless tries until correct answer is inputted
        recruit()

    # End conditional statement for teammate question
      
        
# Define the function of room 3     
def travel():  
    
    # introduce a variable 
    transport = 'ride a whale'
    
    # number of chances to input the correct answer 
    chances   = 2
    
    # start a while loop for with 2 chances to guess 
    while chances > 0:
    
        # print asking users about mode of transportation 
        print(f"""How will you get to Atlantis? 
    1. Book a flight 
    2. Ride a whale
    3. Teleport
""")
    
        # introduce a variable 
        attempts = (input(prompt='>> ')) 
        
        # start a nested conditional statement for transportation question 
        if "2" in attempts or "ride" in attempts or "whale" in attempts or \
        attempts == transport:
            print(f"""\nWhaley great idea! . ><(((.______) 
Take a deep breathe and let's gooooo. 
""")
            # prompt user to continue playing 
            input(prompt='<Press Enter to continue>\n') 
            
            # call for room 4 (name Destination) with correct input 
            destination()
            
            # end while loop once correct answer is inputted 
            break
        
        elif chances != transport:
            
            # user loses a chance when answered wrong 
            chances  -= 1
    
            print(f"""
Sorry, that is currently unavailable. Maybe try something water-related.
You have {chances} chance(s) left.\n""")
        
        else:
            print('Something went wrong.')
            
            # user loses a chance when answered wrong    
            chances  -= 1
          
        # end a nested conditional statement 
    
    # end while loop and call fail function when user run out of chances 
    fail()
  
    
# Define the function of room 4 
def destination():
    
    # print instructions specifying users must answer 2 questions to win 
    print(f"""
    
┌─────────────── ⋆⋅☆⋅⋆ ──────────────┐
  TO WIN: PASS SECURITY CHECK POINTS
└─────────────── ⋆⋅☆⋅⋆ ──────────────┘    

Welcome to Atlantis Customs and Border Protection! ᕦ(⌐■_■)
You have ONE CHANCE to enter. Entry is contingent on providing TWO CORRECT 
ANSWERS IN A ROW. If you fail on the first attempt, your access will 
be denied immediately.
""")
    
    # prompt users to continue playing 
    input(prompt='<Press Enter to continue>\n') 
    
    print(f"""Whales are Atlantisians' guardians. Due to a lingering 
pandemic, they have socially distanced themselves to protect all civilians.
""")
    # time delay for better story-telling and user experience 
    time.sleep(6)
    
    # print 1st in room 4 
    print(f"""Your question is: What do whales do during quarantine?
    1. Netflix and Krill 
    2. No idea
""")
    
    # propmt user to enter the answer for 1st question in room 4 
    # introduce a variable 
    response = (input(prompt='>> '))
    
    # control user input by returning a lowercased or original string
    response = response.lower()
    
    # start a conditional statement for 1st question in room 4 
    if "1" in response or "netflix" in response or "krill" in response:
        print(f"""\nCorrect! 
Ha ha ha ... get it? Because whales eat krill, and ... *sigh*  
just a sneaky pundemic joke to lighten up your day! 
""")
        
        # prompt users to continue playing 
        input(prompt='<Press Enter to continue>\n') 
        
        # print the 2nd/ final question in room 4 with hints to ensure wins 
        print(f"""\nWE ARE ALMOST THERE! 
In front of you is the gate to Atlantis. But it's locked. 
Using what you know, enter the passcode.

Hint: It rhymes with 'Netflix and Chill'
""")
        
        # number of chances to input the correct answer 
        guess = 1
        
        # start while loop for with 1 chance to guess 
        while guess >= 0:
            
            # introduce a variable 
            passcode = input(prompt='>> ')
            
            # control user input by returning a lowercased or original string
            passcode = passcode.lower()
            
            # introduce a variable 
            unlock   = 'netflix and krill'
        
            # start a nested conditional statement for 2nd question in room 4 
            if passcode == unlock: 
                print(f"""
╔══════ ≪ °❈° ≫ ══════╗
  PERMISSION GRANTED
╚══════ ≪ °❈° ≫ ══════╝
""")
                
                # time delay for suspense 
                time.sleep(3)
                
                # print victory message 
                print(f"""
            
            
               __   __
              __ \ / __
             /  \ | /  \
                 \|/
            _,.---v---._
   /\__/\  /            \        ♪ღ♪*•..¸¸.•*¨¨*•.♪ღ♪*•.¸¸.•*¨¨*•..♪ღ♪
   \_  _/ /              \       ░ W ░ H ░ A ░ L ░ E ░ C ░ O ░ M ░ E ░
     \ \_|           @ __|       ♪ღ♪*•..¸¸.•*¨¨*•.♪ღ♪*•.¸¸.•*¨¨*•..♪ღ♪
      \                \_
       \     ,__/       /
     ~~~`~~~~~~~~~~~~~~/~~~~             
                            
                           WHALE-COME TO ATLANTIS!
                Soak in the vitamin sea and enjoy your stay!
              

""")
                # call exit function to quit the game 
                exit()
            
            else:
                if guess  != 0:
                    
                    # user loses a chance when answered wrong 
                    guess -= 1
                    
                    # indirectly prompt users to try again 
                    print(f"""\nAccess denied. 
Ok, we lied. You actually get another try... but only ONE more try.
""")
                    # prompt users to try again with reminder of last try 
                    input('<Press Enter to try again ONE LAST TIME.>\n') 
                
                else:
                    print("\nAccess denied.\n")
                    
                    # call fail function when user run out of chances 
                    fail()
            
            # end a nested conditional statement 
    
    else: 
        print("\nAccess denied.\n")
        
        # call fail function when user answer wrong on first try 
        fail()
        
        # end a while loop 
        
    # end a conditional statement 
    

# Define the function of fail    
def fail():
    
    # print to encourage user to replay the game 
    print(f"""
Aw, you were so close...
    
Don't let your dreams just be dreams. Would you like to try again?
    1. Yes
    2. No
""")
    
    # prompt user to input 'yes' or 'no'
    play = (input(prompt='>> '))
    
    # control user input by returning a lowercased or original string
    play = play.lower()
    
    # start a conditional statement for replay 
    if "1" in play or play == 'yes':
        
        # initiate replay by calling room 1 
        intro()
    
    else:
        print(f"""\nAlrighty, thank you for participating in our program! 

    ＼（＾∀＾）メ（＾∀＾）ノ

Enjoy the rest of your day! 
""")
    
    # call exit function to quit the game 
    exit() 
    

# call to start the game
start()
```
